```python
words = word_tokenize(text)

print(len(words))

words
```

```python
from nltk.probability import FreqDist

fdist = FreqDist(words)

fdist.most_common(10)
```

```python
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
# notice the "," and "." 

# Let's remove punctuation marks and make everything lowercase

cleaned_words = []

for w in words:
    if w.isalpha():
        cleaned_words.append(w.lower())

print (cleaned_words)
```

```python
import textwrap

wrapper = textwrap.TextWrapper(width=50) 
word_list = wrapper.wrap(text=' '.join(cleaned_words))
word_list
```

```python
fdist = FreqDist(cleaned_words)
fdist.most_common(10)
```

```python
fdist.plot(25)
```

```python
# we still have a lot of not very useful works like "and" 
# https://gist.github.com/sebleier/554280

from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words("english")
wrapper.wrap(text=' '.join(stopwords))
```

```python
# let's remove stopwords

without_stopwords = []

for w in cleaned_words:
    if w not in stopwords:
        without_stopwords.append(w)

wrapper.wrap(text=' '.join(without_stopwords))
```

```python
# Final Frequency Distribution
fdist = FreqDist(without_stopwords)
fdist.most_common(10)
```

```python
fdist.plot(20)
```

```python
# stemming is the process of reducing inflected (or sometimes derived) 
# words to their word stem, base or root form—generally a written word form. .
```

```python
# Stemming normalizes the word by truncating the word to its stem word. 
# For example, the words “studies,” “studied,” “studying” will be 
# reduced to “studi,” making all these word forms to refer to only 
# one token. #
```

```python
# Porter's Stemmer 

from nltk.stem import PorterStemmer

porter = PorterStemmer()

for w in without_stopwords:
    print(porter.stem(w))
```

```python
# yes, words are not always recognizable
```

```python
# Lemmatization tries to achieve a similar base “stem” for a word. 
# However, what makes it different is that it finds the dictionary 
# word instead of truncating the original word. Stemming does not 
# consider the context of the word. That is why it generates results 
# faster, but it is less accurate than lemmatization.
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("vehicles"), porter.stem("vehicles")
```

```python
!wget https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-no-swears.txt
```

```python
with open("google-10000-english-no-swears.txt") as f:
    passage = f.read()
    words = passage.lower().split('\n')
print(f'No of words: {len(words)}')
print(f'Some of the words: {words[450:460]}')
```

```python
rank_list = [rank for rank, _ in enumerate(words)]
```

```python
# Encode rank as one-hot vectors

!pip install torch
```

```python
import torch
```

```python
one_hot_vec = torch.zeros(len(words), len(words))

torch.arange(9895)
```

```python
len(words)
```

```python
rank_list
```

```python
one_hot_vec[torch.arange(len(words)), rank_list] = 1
```

```python
print(f'One hot vector size: {one_hot_vec.shape}')
```

```python
import sys
sys.getsizeof(one_hot_vec.storage())
```

```python
# Encode rank as embeddings

embeds = torch.nn.Embedding(len(words), 128)
rank_tensor = torch.LongTensor(rank_list)
embeddings = embeds(rank_tensor)

print(f'One hot vector size: {embeddings.shape}')
```

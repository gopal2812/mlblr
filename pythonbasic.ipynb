```python
import numpy as np

def quicksort(eip):
    if len(eip) <= 1:
        return eip
    eip_out = eip[len(eip) // 2]
    mlblr = [eip_in for eip_in in eip if eip_in < eip_out]
    mlblr_in = [eip_in for eip_in in eip if eip_in == eip_out]
    mlblr_out = [eip_in for eip_in in eip if eip_in > eip_out]
    return quicksort(mlblr) + mlblr_in + quicksort(mlblr_out)
print(quicksort([3,6,8,10,1,2,1]))
```

```python
def eip_op(eip, mlblr=False):
    if mlblr:
        print('HELLO, %s!' % eip.upper())
    else:
        print('Hello, %s' % eip)

eip_op('Bob') # Prints "Hello, Bob"
eip_op('Fred', mlblr=True)  # Prints "HELLO, FRED!"
```

```python
class eip(object):

    # Constructor
    def __init__(self, eip_in):
        eip.eip_in = eip_in  # Create an instance variable

    # Instance method
    def mlblr(self, mlblr_in=False):
        if mlblr_in:
            print('HELLO, %s!' % self.eip_in.upper())
        else:
            print('Hello, %s' % self.eip_in)

eip_out = eip('Fred')  # Construct an instance of the Greeter class
eip_out.mlblr()            # Call an instance method; prints "Hello, Fred"
eip_out.mlblr(mlblr_in=True)   # Call an instance method; prints "HELLO, FRED!"
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

eip = np.array([[0, 1], [1, 0], [2, 0]])
print(eip)

# Compute the Euclidean distance between all rows of x.
# d[i, j] is the Euclidean distance between x[i, :] and x[j, :],
# and d is the following array:
# [[ 0.          1.41421356  2.23606798]
#  [ 1.41421356  0.          1.        ]
#  [ 2.23606798  1.          0.        ]]
eip_op = squareform(pdist(eip, 'euclidean'))
print(eip_op)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
eip = np.arange(0, 3 * np.pi, 0.1)
mlblr_in = np.sin(eip)
mlblr_out = np.cos(eip)

# Plot the points using matplotlib
plt.plot(eip, mlblr_in)
plt.plot(eip, mlblr_out)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
eip = np.arange(0, 3 * np.pi, 0.1)
mlblr_in = np.sin(eip)
mlblr_out = np.cos(eip)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(eip, mlblr_in)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(eip, mlblr_out)
plt.title('Cosine')

# Show the figure.
plt.show()
```
